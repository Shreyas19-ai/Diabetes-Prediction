```python
import scipy.stats as stats
import pylab

def plot_data(df, feature):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    axes[0].hist(df[feature], color='blue', alpha=0.7, bins=30)
    axes[0].set_title(f"Histogram of {feature}")
    axes[0].set_xlabel(feature)
    axes[0].set_ylabel("Frequency")

    stats.probplot(df[feature], dist='norm', plot=axes[1])
    axes[1].set_title(f"QQ Plot of {feature}")

    plt.tight_layout()
    plt.show()
```

```python
plot_data(df, 'BloodPressure')
plot_data(df, 'Insulin')
plot_data(df, 'Glucose')
plot_data(df, 'Age')
plot_data(df, 'SkinThickness')
plot_data(df, 'BMI')
```

```python
X = df.drop(['Outcome'], axis = 1)
y = df['Outcome']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
from imblearn import SMOTE
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
dt1 = DecisionTreeClassifier()
dt1.fit(X_train, y_train)
```

```python
plt.Figure(figsize=(15,10))
tree.plot_tree(dt1, filled = True)
```

```python
parameter = {'criterion': ['gini','entropy','log_loss'],
             'max_depth': [1,2,3,4,5,6,7],
             'splitter': ['best', 'random'],
             'max_features': ['auto', 'sqrt', 'log2']}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
```

```python
dt = DecisionTreeClassifier()
cv = GridSearchCV(dt, param_grid = parameter, cv = 5, scoring = 'accuracy')
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
```

```python
np.array(y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(acc)
print(precision)
print(recall)
```

```python
print(classification_report(y_test, y_pred))
```
